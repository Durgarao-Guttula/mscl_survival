## FFT for background subtraction

A better approach for background subtraction is to use a FFT approach with masking rather than just computing the mean of the dilated negative segmentation mask. Here, I work on a scheme for efficient and simple implementation while preserving quantitative information. 

```python
# Load and example image. 
data_dir = 'data/images/'
gfp_im = skimage.io.imread(data_dir + 'sd0/Jul14/SD0_pre_3/Pos0/img_000000000_GFP_000.tif')

# Show the image in log scale so it's more obvious to the eye. 
plt.imshow(np.log(gfp_im), cmap='viridis')
```

```python
fft = np.fft.fft2(gfp_im)
fshift = np.fft.fftshift(fft)
power_spectrum = np.log(np.abs(fshift)**2)
mask = power_spectrum < 18 
conv = mask * power_spectrum
shift = np.fft.ifftshift(conv)
rev = np.fft.ifft2(shift)
plt.imshow(np.abs(rev), cmap='viridis')
plt.colorbar()

```
