## Making an example plot. 

```python
props = {'background_fill_color': '#E3DCD0',
        'outline_line_color' : None}
```

```python
def bokeh_boiler():
   # Make a bokeh figure axis. 
    p = bokeh.plotting.figure()
                             
    p.background_fill_color='#E3DCD0'
    p.xgrid.grid_line_color = '#FFFFFF'
    p.ygrid.grid_line_color = '#FFFFFF'
    p.xgrid.grid_line_dash = 'dotted'
    p.ygrid.grid_line_dash = 'dotted'
    p.xgrid.grid_line_width = 2
    p.ygrid.grid_line_width = 2
    p.axis.minor_tick_line_color = None
    p.axis.major_tick_line_color = None
    p.axis.axis_line_color = None
    p.axis.axis_label_text_font = 'Lucida Sans Unicode'
    p.axis.axis_label_text_font_style = 'normal'
    p.axis.axis_label_text_font_size = '13pt'
    p.axis.axis_label_text_color = '#3c3c3c'
    p.axis.axis_label_standoff = 2
    return p

 
```

```python
# Make a bokeh figure axis. 

p1 = bokeh_boiler()
p2 = bokeh_boiler()
p3 = bokeh_boiler()

# Now for the actual plotting. 
x = np.random.normal(size=5000)
y = np.random.normal(size=5000)

p1.circle(x, y, size=4, color='red', alpha=0.5)
p2.circle(x, y, size=4, color='green', alpha=0.5)
p3.circle(x, y, size=4, color='dodgerblue', alpha=0.5)


p1.xaxis.axis_label= 'log summed intensity '
p2.yaxis.axis_label= 'log squared difference'


# grid = bokeh.layouts.gridplot([[p1, None], [p2, p3]])
bokeh.io.show(p1)
# Specify and plot the glyhs
```
